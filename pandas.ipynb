```python
#Group By
data1 = {'Name':['Jai','Anuj','Jai','Prince','Gaurav','Anuj','Prince','Abhi','Gaurav'],
          'Age':[27,24,22,32,33,36,27,32,28],
           'Address':['Nagpur','Kanpur','Allahabad','Kannauj','Jaunpur','Kanpur','Allahabad','Aligarh','Varanasi'],
         'Qualification':['Msc','MA','MCA','Phd','B.tech','B.com','Msc','MA','B.tech'],
         'Salary': [1000,2000,3000,4000,5000,6000,7000,8000,9000]
        }
print(type(data1))

```

```python
df_new = pd.DataFrame(data1)
df_new
```

```python
#Group by column name
df_new.groupby('Name')
print(df_new.groupby('Name').groups)
```

```python
df2 = df_new.groupby("Name")
list(df2)
```

```python
for i in df2:
    print (i)
```

```python
#groupby using multiple keys
x = df_new.groupby(["Name","Qualification"])
print(x.groups)
```

```python
df_new.groupby("Name")
print(df_new.groupby("Name").sum())
```

```python
df_new.groupby("Address")
print(df_new.groupby("Address").sum())
```

```python
#Merging of two data-frames
data1 = {'Name':['Jai','Anuj','Jai','Prince'],
           'Age':[27,24,22,32],
           'Address':['Nagpur','Kanpur','Allahabad','Kannauj'],
         'Qualification':['Msc','MA','MCA','Phd'],
         'Salary': [1000,2000,3000,4000]
        }
data2 = {'Name':['Gaurav','Anuj','Prince','Abhi','Gaurav'],
          'Age':[33,36,27,32,28],
           'Address':['Jaunpur','Kanpur','Allahabad','Aligarh','Varanasi'],
         'Qualification':['B.tech','B.com','Msc','MA','B.tech'],
         'Salary': [5000,6000,7000,8000,9000]
        }
```

```python
#Converting dict to df
df1 = pd.DataFrame(data1,index = [0,1,2,3])
df2 = pd.DataFrame(data2,index = [4,5,6,7,8])
```

```python
#Merging the dataframes
df_final = [df1,df2]
res = pd.concat(df_final)
print(res)
```
